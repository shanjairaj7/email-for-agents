## Setup

Set your API keys. In Google Colab, use **Secrets** (lock icon in the sidebar) to store `COMMUNE_API_KEY` and `OPENAI_API_KEY` securely.

````python
import os
from commune import CommuneClient
from langchain_openai import ChatOpenAI

COMMUNE_API_KEY = os.environ.get("COMMUNE_API_KEY", "comm_your_key_here")
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-your_key_here")

commune = CommuneClient(api_key=COMMUNE_API_KEY)
llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0.3)

# Create a dedicated support inbox
inbox = commune.inboxes.create(local_part="support-agent")
print(f"\u2713 Support inbox ready: {inbox.address}")
print(f"  Inbox ID: {inbox.id}")
````

## Define Commune Tools for LangChain

We wrap the Commune SDK operations as `@tool`-decorated functions. LangChain's agent will decide which tools to call based on the current task.

The four tools below give the agent full read/write access to the inbox:
- `send_email` — send a new email or start a thread
- `reply_in_thread` — reply inside an existing conversation
- `read_inbox` — list recent threads
- `search_threads` — semantic search across thread history

````python
from langchain_core.tools import tool

INBOX_ID = inbox.id  # captured from setup above


@tool
def send_email(to: str, subject: str, body: str) -> str:
    """Send a new email from the support inbox to a customer.

    Args:
        to: Recipient email address.
        subject: Email subject line.
        body: Plain-text email body.

    Returns:
        Confirmation string with message ID.
    """
    result = commune.messages.send(
        to=to,
        subject=subject,
        text=body,
        inbox_id=INBOX_ID,
    )
    return f"Email sent. Message ID: {result.get('message_id', 'ok')}"


@tool
def reply_in_thread(thread_id: str, to: str, body: str) -> str:
    """Reply to a customer within an existing email thread.

    Use this tool when responding to an inbound customer email so the reply
    appears in the same conversation thread in the customer's email client.

    Args:
        thread_id: The Commune thread ID (from the webhook payload or threads.list).
        to: Recipient email address (the customer's email).
        body: Plain-text reply body.

    Returns:
        Confirmation string.
    """
    result = commune.messages.send(
        to=to,
        text=body,
        inbox_id=INBOX_ID,
        thread_id=thread_id,
    )
    return f"Reply sent in thread {thread_id}. Message ID: {result.get('message_id', 'ok')}"


@tool
def read_inbox(limit: int = 10) -> str:
    """List the most recent email threads in the support inbox.

    Args:
        limit: Maximum number of threads to return (1-50).

    Returns:
        Formatted list of threads with subject, sender, and message count.
    """
    threads = commune.threads.list(inbox_id=INBOX_ID, limit=limit)
    if not threads.data:
        return "No threads found in inbox."

    lines = []
    for t in threads.data:
        lines.append(
            f"- Thread ID: {t.thread_id} | Subject: {t.subject or '(no subject)'} "
            f"| Messages: {t.message_count} | Last: {t.last_message_at}"
        )
    return "\n".join(lines)


@tool
def search_threads(query: str, limit: int = 5) -> str:
    """Search the support inbox using natural language semantic search.

    Use this to find past threads relevant to a customer's issue. Results are
    ranked by semantic similarity — not keyword matching.

    Args:
        query: Natural language search query, e.g. 'billing issue refund request'.
        limit: Maximum number of results to return.

    Returns:
        Formatted list of matching threads.
    """
    results = commune.search.threads(
        query=query,
        inbox_id=INBOX_ID,
        limit=limit,
    )
    if not results:
        return "No matching threads found."

    lines = []
    for r in results:
        lines.append(
            f"- Thread ID: {r.thread_id} | Subject: {r.subject} "
            f"| Similarity: {r.score:.2f}"
        )
    return "\n".join(lines)


tools = [send_email, reply_in_thread, read_inbox, search_threads]
print(f"\u2713 {len(tools)} tools registered: {[t.name for t in tools]}")
````

## Create the LangChain Support Agent

We use `create_openai_functions_agent` with a system prompt tuned for customer support. The agent can call any of the four tools above during its reasoning loop.

The `AgentExecutor` runs the agent until it reaches a final answer (or hits `max_iterations`).

````python
from langchain.agents import AgentExecutor, create_openai_functions_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

SYSTEM_PROMPT = """You are a helpful customer support agent for a SaaS company.

When you receive a customer email, you must:
1. Classify the issue type: billing, technical, account, general, or feature_request
2. Search for related past threads to understand if this customer has written before
3. Generate an empathetic, helpful reply that addresses the customer's concern
4. Reply in the same thread using reply_in_thread — never open a new thread for an existing conversation

Tone guidelines:
- Warm and empathetic — acknowledge the customer's frustration first
- Specific — reference their actual issue, not a generic template
- Actionable — always end with a clear next step
- Brief — aim for 3-4 sentences maximum

Always use the available tools rather than making up information.
"""

prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_openai_functions_agent(llm=llm, tools=tools, prompt=prompt)
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
    max_iterations=6,
    handle_parsing_errors=True,
)

print("\u2713 LangChain support agent ready")
````

## Webhook Handler (Flask)

In production, Commune POSTs to your webhook URL every time an email arrives. The handler below verifies the signature, extracts the email payload, and runs the agent.

Run this cell to see the full handler code — it won't start a server in this notebook environment.

````python
webhook_handler_code = '''
import json
import os
from flask import Flask, request, Response
from commune import verify_signature, WebhookVerificationError

app = Flask(__name__)

@app.route("/webhook", methods=["POST"])
def handle_inbound_email():
    body = request.get_data()  # raw bytes — required for HMAC verification

    # Step 1: Verify the webhook signature
    try:
        verify_signature(
            payload=body,
            signature=request.headers.get("x-commune-signature", ""),
            secret=os.environ["COMMUNE_WEBHOOK_SECRET"],
            timestamp=request.headers.get("x-commune-timestamp", ""),
        )
    except WebhookVerificationError as e:
        return Response(str(e), status=401)

    # Step 2: Parse the payload
    payload = json.loads(body)
    sender = payload["sender"]
    thread_id = payload["thread_id"]
    subject = payload.get("subject", "(no subject)")
    content = payload["content"]

    # Step 3: Run the LangChain agent
    task = (
        f"New support email received.\n"
        f"From: {sender}\n"
        f"Subject: {subject}\n"
        f"Thread ID: {thread_id}\n"
        f"Message: {content}\n\n"
        f"Please classify this issue, search for any prior context, "
        f"and reply to the customer using reply_in_thread."
    )
    agent_executor.invoke({"input": task})

    # Always return 200 quickly — Commune expects a fast response
    return {"ok": True}

if __name__ == "__main__":
    app.run(port=8000)
'''

print("Webhook handler:")
print(webhook_handler_code)
````

## Common Mistake 1: Missing `thread_id` on Replies

The most frequent LangChain + Commune bug is replying without `thread_id`.

**What goes wrong:** Every call to `messages.send()` without a `thread_id` starts a *brand-new email thread*. The customer opens their inbox and sees five separate unrelated emails instead of one conversation. Email clients like Gmail and Outlook use `In-Reply-To` and `References` headers — which Commune sets automatically when you pass `thread_id` — to group messages. Without them, grouping never happens.

**When it hurts most:** Multi-turn support conversations. The customer replies to email #3; your agent gets a webhook, can't find prior context, and starts email #6 from scratch.


````python
# ❌ WRONG: Calling messages.send() without thread_id for a reply
# Every call here creates a completely new email thread.
# The customer sees a fresh "support@..." email with no connection to their
# original message — no "Re:" chain, no conversation history in Gmail/Outlook.

@tool
def reply_in_thread_broken(thread_id: str, to: str, body: str) -> str:
    result = commune.messages.send(
        to=to,
        text=body,
        inbox_id=INBOX_ID,
        # thread_id is NOT passed — this opens a new thread every single time
    )
    return f"Sent (but in wrong thread!). Message ID: {result.get('message_id')}"
````

````python
# ✅ CORRECT: Always pass thread_id when replying to an existing conversation
# Commune uses thread_id to set the In-Reply-To and References headers,
# which tells Gmail / Outlook to display this as part of the same thread.
# The customer sees one conversation, not five disconnected emails.

@tool
def reply_in_thread(thread_id: str, to: str, body: str) -> str:
    result = commune.messages.send(
        to=to,
        text=body,
        inbox_id=INBOX_ID,
        thread_id=thread_id,   # ← required for conversation continuity
    )
    return f"Reply sent in thread {thread_id}. Message ID: {result.get('message_id')}"

# In your webhook handler, thread_id comes straight from the payload:
#   thread_id = payload["thread_id"]
# In threads.list() results:
#   thread_id = thread.thread_id
````

## Common Mistake 2: Parsing JSON Before Verifying the Signature

Flask and FastAPI both make it tempting to call `request.json()` or `request.get_json()` at the top of the handler. Do that before calling `verify_signature()` and you've broken security.

**What goes wrong:** HMAC is computed over the *raw request bytes*, not the parsed dict. Once you call `request.get_json()`, Flask has already consumed and decoded the body. Any attacker who modifies the JSON in transit gets a structurally valid dict but an invalid signature — except you never checked the signature against raw bytes, so you accept the tampered payload.

**The rule:** Read raw bytes first with `request.get_data()`, verify, *then* call `json.loads()` on those same bytes.


````python
# ❌ WRONG: Parsing JSON before verifying the HMAC signature
# request.get_json() returns a Python dict, but verify_signature() needs the
# original raw bytes that the HMAC was computed over.
# Passing the dict (or re-serialised JSON) will cause every request to fail
# verification — OR worse, you skip verification entirely to "fix" the error.

@app.route("/webhook", methods=["POST"])
def handle_inbound_broken():
    payload = request.get_json()   # ← consumed the body here

    # Now trying to verify with the dict — this will always raise an error
    # because the HMAC was computed over raw bytes, not a Python object.
    try:
        verify_signature(
            payload=payload,                                      # ← wrong type
            signature=request.headers.get("x-commune-signature"),
            secret=os.environ["COMMUNE_WEBHOOK_SECRET"],
            timestamp=request.headers.get("x-commune-timestamp"),
        )
    except Exception:
        pass  # ← common but dangerous: silently ignore verification failures

    sender = payload["sender"]
    # ... rest of handler
````

````python
# ✅ CORRECT: Read raw bytes first, verify, then parse JSON
# The HMAC is computed over the exact bytes Commune sent.
# verify_signature() will raise WebhookVerificationError for any mismatch,
# rejecting spoofed or tampered requests before they touch your agent logic.

@app.route("/webhook", methods=["POST"])
def handle_inbound_email():
    body = request.get_data()   # ← raw bytes — do this FIRST

    try:
        verify_signature(
            payload=body,       # ← bytes, not dict
            signature=request.headers.get("x-commune-signature", ""),
            secret=os.environ["COMMUNE_WEBHOOK_SECRET"],
            timestamp=request.headers.get("x-commune-timestamp", ""),
        )
    except WebhookVerificationError as e:
        return Response(str(e), status=401)  # ← always return 401 on failure

    payload = json.loads(body)  # ← safe to parse now, after verification
    sender     = payload["sender"]
    thread_id  = payload["thread_id"]
    content    = payload["content"]
    # ... run agent ...
    return {"ok": True}
````

## Common Mistake 3: Running the LLM Agent Synchronously in the Webhook Handler

LLM inference takes 2–15 seconds. Commune's webhook delivery expects a response within a few seconds. If your handler blocks on agent_executor.invoke(), you risk timeouts, duplicate deliveries (Commune retries), and a jammed handler that can't accept new emails while one is being processed.

**The failure mode:** Under load, every incoming email queues behind the previous LLM call. Your Flask dev server (single-threaded by default) processes one email at a time. If the LLM takes 8 s and three emails arrive simultaneously, the third webhook times out, Commune retries, and now you have duplicate processing.

**The fix:** Return `{"ok": True}` immediately, push the payload to a queue, and process it in a background worker (Celery, RQ, asyncio task, etc.).


````python
# ❌ WRONG: Blocking the webhook handler with synchronous LLM inference
# agent_executor.invoke() may take 5-15 seconds.
# During that time the handler is blocked — no other emails can be processed.
# If Commune's delivery timeout (~5 s) is shorter than LLM latency, Commune
# retries the webhook. You process the same email twice.

@app.route("/webhook", methods=["POST"])
def handle_inbound_blocking():
    body = request.get_data()
    # (signature verification omitted for brevity)
    payload = json.loads(body)

    # ← blocks here for several seconds — no good under any real load
    result = agent_executor.invoke({"input": build_task(payload)})

    return {"ok": True}  # Commune may never see this if LLM is too slow
````

````python
# ✅ CORRECT: Return 200 immediately, process the email in the background
# The webhook handler just validates, enqueues, and returns.
# A separate worker (thread, Celery task, asyncio coroutine) runs the agent.
# This pattern handles bursts cleanly and avoids duplicate processing.

import threading

def run_agent_async(payload):
    """Called in a background thread — LLM latency doesn't block the handler."""
    task = (
        f"New support email.\nFrom: {payload['sender']}\n"
        f"Subject: {payload.get('subject', '')}\n"
        f"Thread ID: {payload['thread_id']}\n"
        f"Message: {payload['content']}\n\n"
        f"Classify, find prior context, and reply using reply_in_thread."
    )
    agent_executor.invoke({"input": task})


@app.route("/webhook", methods=["POST"])
def handle_inbound_email():
    body = request.get_data()

    try:
        verify_signature(
            payload=body,
            signature=request.headers.get("x-commune-signature", ""),
            secret=os.environ["COMMUNE_WEBHOOK_SECRET"],
            timestamp=request.headers.get("x-commune-timestamp", ""),
        )
    except WebhookVerificationError as e:
        return Response(str(e), status=401)

    payload = json.loads(body)

    # Fire-and-forget: spin up a thread, return 200 immediately
    threading.Thread(target=run_agent_async, args=(payload,), daemon=True).start()

    return {"ok": True}  # ← Commune receives this in milliseconds

# In production, replace threading.Thread with:
#   Celery:  process_email.delay(payload)
#   RQ:      queue.enqueue(run_agent_async, payload)
#   asyncio: background_tasks.add_task(run_agent_async, payload)  (FastAPI)
````
